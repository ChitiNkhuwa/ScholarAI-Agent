```python
import os
from agents import Agent, Runner
from dotenv import load_dotenv

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
```

```python
sarcastic_agent = Agent(
    name="Sarcastic Agent",
    instructions="You are a sarcastic assistant. Respond to all queries with sarcasm.",
    model="gpt-4o-mini"
)
```

```python
result = await Runner.run(sarcastic_agent, "What's the weather like today?")
print(result)
```

```python
result.final_output
```

```python
dummy_stock_agent = Agent(
    name="Dummy Stock Agent",
    instructions="You are a stock market assistant. Provide stock dummy prices when asked.",
    model="gpt-4o-mini",
    handoff_description="You will provide dummy data of a comapnies stock."
)

stock_analyzer_agent = Agent(
    name="Stock Analyzer Agent",
    instructions="You are an expert stock market analyst. Analyze stock trends and provide insights.",
    model="gpt-4o-mini",
    handoff_description="You will analyze stock market trends and provide insights." 
)

triage_agent = Agent(
    name="Triage Agent",
    instructions="You are a triage assistant. Decide whether to use the Dummy Stock Agent or the Stock Analyzer Agent based on the user's query.",
    model="gpt-4o-mini",
    handoffs=[dummy_stock_agent, stock_analyzer_agent]
)
```

```python
result = await Runner.run(triage_agent, "Can you provide the latest stock price for AAPL and analyze its trend?")
```

```python
print(result.final_output)
```
